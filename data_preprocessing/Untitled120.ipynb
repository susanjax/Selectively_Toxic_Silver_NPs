```python
import pandas as pd
```

```python
df_main = pd.read_excel(r'/content/Validation_SelTox_NeurIPS_dasha_checked_15_07_25 (3).xlsx')
df_add = pd.read_excel(r'/content/completed synergy desc+tax (4).xlsx')
```

```python
df_main.head()
```

```python
cols = df_main.columns.append(df_add.columns)
df_temp = pd.DataFrame(columns=cols)
```

```python
df_temp
```

```python
counter = 0
counter_b = 0
data = list()

for index, vals in df_main.iterrows():
  cur_bact, cur_strain = vals['bacteria'], vals['strain']
  current_row_all_same = df_add[(df_add['Bacteria'] == cur_bact) & (df_add['Strain'] == cur_strain)]
  if not current_row_all_same.empty:
    counter_b +=1
    data.append(np.concatenate([vals.values, current_row_all_same.values[0]]))
  else:
    current_row = df_add[(df_add['Bacteria'] == cur_bact) & (df_add['Strain'].isna())]
    if not current_row.empty:
      data.append(np.concatenate([vals.values, current_row.values[0]]))
```

```python
df_temp = pd.DataFrame(data=data, columns=df_main.columns.append(df_add.columns))
```
